````python
quantiles = np.quantile(d, np.linspace(0, 1, 11))
print(quantiles)
''' découper ta variable d en 10 groupes 
 contenant le même nombre d’observations.
C’est une découpe par effectifs égaux, pas par largeur égale.
'''

counts, bins  = np.histogram(d, bins = 30)
print(type(counts))
print(type(bins))
print("Bornes :", bins)
print("Effectifs :", counts)

plt.hist(d, bins = 10)
''' Tu comptes combien de valeurs sont dans chaque intervalle avec 
10 effectifs différents
'''
````

````python
nb_int = 30
borne_min = np.min(d)
print(borne_min)
borne_max = np.max(d)
print(borne_max)
bornes = np.linspace(borne_min, borne_max, nb_int + 1 )
print(bornes)

Eff = []
# np.where trouve les indices qui valident les conditions
indices = np.where((d > bornes[0]) & (d < bornes[1]))[0]
print(len(indices))

for i in range(nb_int) : 
    gau = bornes[i]
    dr = bornes[i+1]

    indices = np.where((d >= gau) & (d < dr))[0]

    Eff.append(len(indices))

Eff

# plt.bar a besoin d'un centre pour dessiner un rectangle
centres = []
for i in range(len(bornes) - 1) :
    c = (bornes[i] + bornes[i+1]) / 2
    centres.append(c)
centres = np.array(centres)
plt.bar(centres, Eff, width=(bornes[1] - bornes[0]))
````

# Distributions jointes, distributions conditionnelles

Nous voulons maintenant étudier la distribution jointe entre la distance et la marque de la voiture. Partir des distributions discrètes ou discétisées et construire le tableau d'effectif puis normaliser par les effectifs de l'échantillon pour estimer la loi jointe.

Il est diffile d'analyser cette probabilité jointe (cf ci-dessous pour l'affichage)... Nous allons donc passer à la loi conditionnelle: nous voulons donc calculer la probabilité de la distance conditionnellement à la marque de la voiture.

1. Proposer un critère rapide pour vérifier que votre distribution conditionnelle respecte bien les propriétés de base
2. Cette distribution conditionnelle fait apparaitre des pics très marqués: pouvons-nous tirer parti de ces informations?

````python
'''
Contexte général

Tu disposes d’un tableau de données data contenant plusieurs colonnes :
→ distance, prix, marque, etc.
Tu veux maintenant étudier statistiquement le lien entre deux variables :

la distance d’un trajet,

et la marque du véhicule.

Autrement dit, tu veux connaître quelle distance typique est associée à quelle marque (et inversement).
'''
titles_col = fich['indexcol']
print(titles_col)
marque = data[:, 11]
print(dico_marques)
print(marque)
print(len(marque))
print(d)
print(len(d))

# nettoyage
dico_inv_marques = {v:k for k, v in dico_marques.items()}
print(dico_inv_marques)
for code in np.unique(data[:, 11]):
    print(code, "→", dico_inv_marques[int(code)])


min_marque, max_marque = np.min(marque), np.max(marque) # pas besoin
min_d, max_d = np.min(d), np.max(d) 
````

````python
print(len(marque))
print(len(d))
# même si les deux vecteurs ont la même longuer, pour la distance, tu dois absolument la discrétiser.

nb_int = 10
nb_d = np.linspace(min_d, max_d, nb_int+1)
print(nb_d)

'''
np.digitize(x, bins) parcourt ton tableau x et renvoie, pour chaque valeur, 
le numéro d’intervalle dans lequel elle se trouve.
'''

d_classes = np.digitize(d, nb_d, right=True) - 1
print(d_classes[:20])
print(d_classes[0:20])

# Allez pour la constuction de la matrice, les distances sur les lignes, vu que augmentent,  elle va sur l’axe vertical (lignes).
nb_classes = len(np.unique(d_classes))
nb_marques = len(np.unique(marque))
print(nb_classes)
print(nb_marques) # nombre de marques différentes
print(len(d_classes)) # nombre total de trajets (échantillons)
print(len(marque)) # nombre total de trajets (échantillons)

# préparer la matrice
p_dm = np.zeros((nb_classes, nb_marques))
plt.imshow(p_dm, interpolation='nearest')


for i in range(len(d_classes)) :
    d_idx = int(d_classes[i])
    m_idx = int(marque[i])
    p_dm[d_idx, m_idx] += 1


plt.imshow(p_dm, interpolation='nearest')
plt.colorbar()
plt.show()


p_dm /= np.sum(p_dm)

# =====================================================================
#  Visualisation de la matrice p_dm (distance x marque)
# =====================================================================

# On crée une figure et un axe sur lesquels on va dessiner le graphique.
# figsize=(10,5) définit la taille de la figure (10 pouces de large, 5 de haut)
fig, ax = plt.subplots(1, 1, figsize=(10, 5))

# Affiche la matrice p_dm comme une image 2D (carte de chaleur)
# interpolation='nearest' empêche le lissage entre les pixels
# la fonction renvoie un objet "image" que l’on stocke dans im (utile pour la colorbar)
im = ax.imshow(p_dm, interpolation='nearest')

# ---------------------------------------------------------------------
# Axe horizontal : les marques
# ---------------------------------------------------------------------

# Définit les positions des étiquettes sur l'axe des X
# np.arange(len(dico_marques)) crée un vecteur [0, 1, 2, ..., nb_marques-1]
ax.set_xticks(np.arange(len(dico_marques)))

# Définit le texte affiché sous chaque graduation (les noms des marques)
# rotation=90 tourne le texte verticalement pour qu’il tienne
# fontsize=8 réduit la taille pour que tout soit lisible
ax.set_xticklabels(dico_marques.keys(), rotation=90, fontsize=8)

# ---------------------------------------------------------------------
# Axe vertical : les classes de distance
# ---------------------------------------------------------------------

# Ajoute un label pour l’axe Y
ax.set_ylabel("Classes de distance")

# Ajoute un label pour l’axe X
ax.set_xlabel("Marques")

# ---------------------------------------------------------------------
# Échelle de couleurs (colorbar)
# ---------------------------------------------------------------------

# Ajoute une barre de couleur à droite de la figure pour indiquer
# la valeur associée à chaque nuance (plus clair = plus de trajets)
# label=... permet d’ajouter une légende à cette barre
fig.colorbar(im, ax=ax, label="Nombre de trajets")

# ---------------------------------------------------------------------
# Mise en page et sauvegarde
# ---------------------------------------------------------------------

# Ajuste automatiquement la mise en page pour éviter que les textes
# soient coupés (notamment les marques en bas)
plt.tight_layout()

# # Sauvegarde la figure au format PDF (image vectorielle : qualité infinie au zoom)
# plt.savefig("mafigure.pdf")

# Affiche la figure à l'écran
plt.show()


````

````python
# Loi Marginale
P_marque = p_dm.sum(axis=0)
print(P_marque)
print(P_marque.shape)
print("Somme totale :", np.sum(P_marque))
print(dico_marques)
print(dico_inv_marques)
for code, prob in enumerate(P_marque):
    marque_nom = dico_inv_marques.get(code, f"Code {code}")
    print(f"{marque_nom:<15} : {prob:.4f}") # :.4f pour limiter à 4 chiffres
indices_trie = np.argsort(P_marque)
top = indices_trie[-1] # trie du plus petit au plus grand
print(dico_inv_marques[top], P_marque[top])

````

````python
# Loi Conditionnelle

P_marque = p_dm.sum(axis=0, keepdims=True)

cond_dm = np.divide(
    p_dm, P_marque,                         # division élément par élément
    out=np.zeros_like(p_dm),                # si division impossible → 0
    where=P_marque > 0                      # évite /0 si marque absente
)

plt.imshow(cond_dm)
print(cond_dm.sum(axis=0))
print(cond_dm.sum())  
print(cond_dm.shape)
print(p_dm.shape)
print(P_marque.shape)
````
